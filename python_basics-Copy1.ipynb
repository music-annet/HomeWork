Пример работы программы:

```
Введите день:
30

```
Результат:

```

29

Введите месяц:
Октябрь
```
Результат:  
`Ваш знак зодиака: Скорпион` 


````python
day = int(input('Введите день: '))
month=input('Введите месяц: ')
print('Ваш знак зодиака:')
if month=="Март":
    if day<=21:
        print('Рыбы')
    else:
        print('Овен')
elif month=="Апрель":
    if day<=20:
        print('Овен')
    else:
        print('Телец')
elif month=="Май":
    if day<=21:
        print('Телец')
    else:
        print('Близнецы')
elif month=="Июнь":
    if day<=21:
        print('Близнецы')
    else:
        print('Рак')
elif month=="Июль":
    if day<=22:
        print('Рак')
    else:
        print('Лев')
elif month=="Август":
    if day<=22:
        print('Лев')
    else:
        print('Дева')
elif month=="Сентябрь":
    if day<=22:
        print('Дева')
    else:
        print('Весы')
elif month=="Октябрь":
    if day<=23:
        print('Весы')
    else:
        print('Скорпион')
elif month=="Ноябрь":
    if day<=22:
        print('Скорпион')
    else:
        print('Стрелец')
elif month=="Декабрь":
    if day<=21:
        print('Стрелец')
    else:
        print('Козерог')
elif month=="Январь":
    if day<=20:
        print('Козерог')
    else:
        print('Водолей')
elif month=="Февраль":
    if day<=18:
        print('Водолей')
    else:
        print('Рыбы')
else :
    print('Проверьте правильно ли вы вели все!')
````

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`

## Задание 6

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
type=input('Введите тип фигуры:')
if type=="Круг" :
    R=int(input('Введите радиус круга:'))
    S=3.14*R*R
    print('Площадь круга:', S)
elif type=="Треугольник" :
    A=int(input('Введите длину стороны A:'))
    B=int(input('Введите длину стороны B:'))
    C=int(input('Введите длину стороны C:'))
    p=(A+B+C)/2
    S=(p*(p-A)*(p-B)*(p-C))**0.5
    print('Площадь треугольника:', S)
elif type=="Прямоугольник" :
    A=int(input('Введите длину стороны A:'))
    B=int(input('Введите длину стороны B:'))
    S=A*B
    print('Площадь прямоугольника:', S)
else :
     print('Такой фигуры нет!')
````
